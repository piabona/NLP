# 자연어란?
- 사람들이 일상적으로 쓰는 언어를 인공적으로 만들어진 언어인 인공어와 구분하여 부르는 개념
- 인류가 살아가는 과정에서 자연스레 만들어졌기 때문에 자연어라고 함
- 언어학 관점에서 언어는 자연어, 인공어, 형식어 등이 있음
    - 자연어 : 한국어, 영어, 일본어, 중국어 등
    - 인공어 : 언어 전문가가 제작한 언어 (반지의 제왕, 아바타, 스타워즈 등 영화속 언어 및 에스페란토)​
    - 형식어 : 수식, 프로그래밍 언어 등


# 텍스트 분석(Text Analytics)
- 텍스트 데이터에서 언어학, 통계학, 기계 학습의 기법을 이용하여 의사결정, 조사, 연구 등의 목적으로 정보를 추출하고 분석하는 것
- 자연어 처리와 많은 부분을 공유

# 자연어 처리(Natural Language Processing)
- 언어학, 컴퓨터 과학, 인공지능 등의 한 분야로서 컴퓨터를 이용해 자연어를 자동으로 처리하는 방법을 연구
- 기법과 내용에서 텍스트 분석과 많은 부분을 공유

## NLU(Natural Language Understanding)
- 자연어 형태의 문장을 이해하는 기술    
- 자연어 -> 정보


## NLG(Natural Language Generation)
- 자연어 문장을 생성하는 기술
- 정보 -> 자연어 텍스트


## 자연어 처리 응용
- 문서 분류
    - 문서를 미리 정해진 유형으로 분류하는 것
    - 뉴스기사 분류, 논문 분류 등
- 감성 분석
    - 문장에 나타난 긍/부정의 태도를 분류
    - 영화리뷰를 읽어서 긍정인지 부정인지
- 개체명 인식
    - 사람, 장소, 단체 등 특정한 종류의 이름(개체명)을 인식
    - 텍스트에서 특정한 종류의 정보를 추출하고자 할 때

- data to text
    - 기사, 보고서 등 생성
- image to text
    - OCR(광학 문자 인식), 이미지 캡션 생성 등

- text to text
    - 기계번역, 챗봇, Q&A, 요약
    - 요약은 긴 텍스트를 짧은 텍스트로 요약

- text to image
    - 입력된 자연어 설명과 일치하는 이미지 생성


## 자연어 처리 어려움
- 해석 가능성​
    - 예측이나 분석결과, 생성된 원리를 해석하기 어려움​

- 데이터 부족​
     - 지도학습의 경우 label이 있어야 함. (label 이 있는 데이터 부족)


- 편향​
     - 데이터 편향 -> 학습한 패턴도 편향 (예측도 편향)


- 신조어​
     - 새로운 단어의 출현시 처리 방법 부재

- 모호성​

    - 자연어 데이터에는 거의 항상 모호성이 존재함​
    - 동의이의어, 문장내 정보 부족​, 상황의 서로 다른 묘사​


- 한국어의 어려움​
    - 교착어 : 어근에 접사가 결합하여 의미가 변화하는 형태의 언어
    - 자유도 높은 띄어쓰기 : 띄어쓰지 않아도 문맥의 이해가 가능​
    - 동일한 평서문과 의문문 : 밥 먹었어. <-> 밥 먹었어?

# 텍스트 수집
- 문서 파일 / 데이터베이스
- 웹 스크래핑
- 음성 인식
- 광학 문자 인식(Optical Character Recognition)
- 수집된 데이터를 말뭉치(corpus) 라고 한다.
    - corpus: NLP 모델에 학습시키기 위한 데이터를 의미



# 텍스트 전처리(Text Preprocessing)
- 불필요한 표현 삭제하고 분석에 편리한 형태로 변환


## 대소문자 정리
- 소문자로 처리할 때 주의할 점으로 소문자로 변경시 뜻이 달라지면 변경하지 말하야 함
    - Language -> language  변경시 뜻이 같음
    - US -> us 뜻이 달라짐


## 데이터 사용 목적에 맞추어 노이즈를 제거
- 불필요한 문장부호, 특수문자 등 삭제
- 불용어(Stop word) 제거
- 출현 횟수가 적은 단어의 제거  




## Stemming(어간 추출)
- 어간을 분리해내는 것
- 어간(Stem): 단어의 의미를 담은 핵심
    - playing -> play(어간)
    - lectures -> lecture(어간)
    - kindness -> kind(어간)

## Lemmatization(표제어 추출)
- 사전에 등재된 형태(lemma)로 바꾸는 것
- lemmatization는 문장에서 단어의 원형을 추출하는 과정
    - is, are -> be
    - having -> have


## 어간 추출 vs 표제어 추출
- 표제어 추출은 단어의 품사 정보를 포함하고 있음
- 어간 추출은 품사 정보를 갖고 있지 않음



## Tokenization(토큰화)
- 자연어 처리를 위해 문장을 처리 단위인 토큰(token)으로 분리하는 것
- 토큰(token): 텍스트 분석의 단위 (단어, 형태소 등등)




## Vocabulary(어휘집)
- 중복을 제거한 어휘와 index가 정의된 집합이다.
- 어휘집을 통해 문자를 숫자로 or 숫자를 문자로 변환


## Padding(패딩)
- 문장의 길이가 달라 같은 길이로 만드는 과정​
    - 긴 경우 : 자르기 (truncate)
    - 짧은 경우 : 채우기 (fill)

## 원-핫 인코딩(One-Hot Encoding)
- 원-핫 인코딩을 통해서 나온 원-핫 벡터들은 표현하고자 하는 단어의 인덱스의 값만 1이고, 나머지 인덱스에는 전부 0으로 표현되는 벡터 표현 방법

- 단어들을 의미있는 실수 형태의 벡터로 표현하는 것
- 신경망에서 임베딩 레이어를 추가하여 단어에 임베딩을 얻을 수 있음
